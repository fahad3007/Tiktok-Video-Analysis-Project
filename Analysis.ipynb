```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Objectdetection.csv')
print(df)
```

```python
feature_counts1 = df['Object 1'].value_counts().rename_axis('Object 1').reset_index(name='count')
feature_counts2 = df['Object 2'].value_counts().rename_axis('Object 2').reset_index(name='count')
feature_counts3 = df['Object 3'].value_counts().rename_axis('Object 3').reset_index(name='count')
```

```python
print(feature_counts1)
```

```python
fig, ax = plt.subplots()
ax.bar(feature_counts1['Object 1'], feature_counts1['count'])

# Customize the plot with axis labels and a title
ax.set_xlabel('Dominant Object')
ax.set_ylabel('Count')
ax.set_title('Most Dominant Object per video')

# Rotate the x-axis labels to make them more readable
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(feature_counts2['Object 2'], feature_counts2['count'])

# Customize the plot with axis labels and a title
ax.set_xlabel('Object')
ax.set_ylabel('Count')
ax.set_title('2nd most Dominant Object per video')

# Rotate the x-axis labels to make them more readable
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(feature_counts3['Object 3'], feature_counts3['count'])

# Customize the plot with axis labels and a title
ax.set_xlabel('Object')
ax.set_ylabel('Count')
ax.set_title('3nd most Dominant Object per video')

# Rotate the x-axis labels to make them more readable
plt.xticks(rotation=90)

# Display the plot
plt.show()
```
